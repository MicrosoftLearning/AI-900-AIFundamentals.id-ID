## Membuat Sumber Daya Prediksi dan Penulisan

Layanan kognitif Microsoft mencakup layanan Pemahaman Bahasa yang memungkinkan Anda menentukan *maksud* yang diterapkan ke *entitas* berdasarkan *ucapan* 

Untuk menggunakan layanan Pemahaman Bahasa, Anda memerlukan dua jenis sumber daya:

- Sumber daya *penulisan*: digunakan untuk menentukan, melatih, dan menguji model bahasa. Hal ini harus merupakan sumber daya **Pemahaman Bahasa - Penulisan** di langganan Azure Anda.
- Sumber daya *prediksi*: digunakan untuk menerbitkan model dan menangani permintaan dari aplikasi klien yang menggunakannya. Hal ini dapat berupa sumber daya **Pemahaman Bahasa** atau **Cognitive Services** di langganan Azure Anda.

Anda dapat menggunakan sumber daya **Pemahaman Bahasa** atau **Cognitive Services** untuk *menerbitkan* aplikasi Pemahaman Bahasa, tetapi Anda harus membuat sumber daya **Pemahaman Bahasa** terpisah untuk *penulisan* aplikasi.

Lihat [dokumentasi wilayah penulisan dan penerbitan](https://docs.microsoft.com/azure/cognitive-services/luis/luis-reference-regions) untuk detail tentang pencocokan lokasi penulisan dan prediksi.

### Membuat Entitas

*Entitas* adalah sesuatu yang dapat diidentifikasi dan dilakukan oleh model bahasa. Dalam hal ini, aplikasi Pemahaman Bahasa Anda akan digunakan untuk mengontrol berbagai *perangkat* di kantor, seperti lampu atau kipas; maka Anda akan membuat entitas *perangkat* yang mencakup daftar jenis perangkat yang Anda inginkan untuk dapat bekerja dengan aplikasi. Untuk setiap jenis perangkat, Anda akan membuat subdaftar yang mengidentifikasi nama perangkat tersebut (misalnya *penerangan*) dan sinonim apa pun yang dapat digunakan untuk merujuk ke jenis perangkat ini (misalnya *lampu*).

1. Di halaman Pemahaman Bahasa untuk aplikasi, di panel sebelah kiri, klik **Entitas**. Lalu klik **Buat**, dan buat entitas baru bernama **perangkat**, pilih jenis **Daftar**, dan klik **Buat**.
2. Di halaman **Daftar item**, di bagian **Nilai yang Dinormalisasi**, ketik **lampu**, lalu tekan ENTER.
3. Setelah nilai **penerangan** ditambahkan, di bagian **Sinonim**, ketik **lampu** dan tekan ENTER.
4. Tambahkan daftar item kedua bernama **kipas** dengan sinonim **AC**.

> **Catatan**: Untuk lab ini, gunakan teks huruf kecil atau huruf besar yang tepat seperti yang telah diinstruksikan _(misal: penerangan **bukan** Penerangan)_ dan jangan tambahkan spasi tambahan. 

### Membuat Maksud

Untuk setiap maksud, Anda akan menentukan sampel *ucapan* yang mengindikasikan jenis bahasa yang digunakan untuk mengindikasikan maksud tersebut.

> **Catatan**: Untuk lab ini, gunakan teks huruf kecil atau huruf besar seperti yang telah diinstruksikan _(example: "nyalakan lampu" **bukan** "Nyalakan penerangan.")_ dan jangan tambahkan spasi tambahan.

3. Dengan ucapan *nyalakan lampu*, klik kata "lampu", dan tetapkan ke nilai **lampu** dan entitas **perangkat**.

Sekarang jalankan sel di bawah untuk mentranskripsikan ucapan dari file audio, dan gunakan sebagai perintah untuk aplikasi Pemahaman Bahasa Anda.

```python
import os
from python_code import luis
from azure.cognitiveservices.speech import SpeechConfig, SpeechRecognizer, AudioConfig
from playsound import playsound
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline

try:   

    # Get spoken command from audio file
    file_name = 'light-on.wav'
    audio_file = os.path.join('data', 'luis', file_name)

    # Configure speech recognizer
    speech_config = SpeechConfig(cog_key, cog_location)
    audio_config = AudioConfig(filename=audio_file) # Use file instead of default (microphone)
    speech_recognizer = SpeechRecognizer(speech_config, audio_config)

    # Use a one-time, synchronous call to transcribe the speech
    speech = speech_recognizer.recognize_once()

    # Get the predicted intent and entity (code in python_code.home_auto.py)
    action = luis.get_intent(luis_app_id, luis_key, luis_endpoint, speech.text)

    # Get the appropriate image
    img_name = action + '.jpg'

    # Display image 
    img = Image.open(os.path.join("data", "luis" ,img_name))
    plt.axis('off')
    plt. imshow(img)
    playsound(audio_file)

except Exception as ex:
    print(ex)
```
